```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
